## Le cycle de vie d un courbe sur matplotlib

```python
'''dans mathplot lib en commance par la fonction figure il ne permer de cree
une feuille blanche ou en peux ecrire notre plot.
'''
```

```python
plt.figure() # entre parenteses vous pouvez ajouter taille de figure
# cree des courbes
x = np.linspace(0,2,10)
y1 = x**5
y2 = 2*x**3+3
y3 = 10*x
plt.plot(x,y1,label = 'courbe1')     # la 1er courbe
plt.plot(x,y2,label = 'courbe2')     # 2 eme courbe
plt.plot(x,y3,label = 'courbe3')     # 3 eme courbe
plt.title('fig 1')                   # nomer la figure 
plt.xlabel('X')                      # nomer l'axe des X
plt.ylabel('Y')                      # nomer l'axe des y
plt.legend()                         # sert a cree une legend    
plt.show()                           # pour afficher reelement la fig
plt.savefig('image.png')             # pou enregistre la fig dans notre repertoire de travaille 
```

## une grile de defferent graphique

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x,np.sqrt(x),c='blue')
plt.subplot(2,2,2)
plt.plot(x,y,c='black')
plt.subplot(2,2,3)
plt.plot(x,np.sqrt(x),c='yellow')
plt.subplot(2,2,4)
plt.plot(x,y3,c='red')
plt.show()
```

# II. METHODE 2 : OOP 
_(programation orionter object)_


```python
fig, ax = plt.subplots(2,2,sharey=True)
ax[0][0].plot(x,y)
ax[0][1].plot(x,np.sin(x))
ax[1][0].plot(x,y)
ax[1][1].plot(x,np.sin(x))
plt.show()
```
